### Género más popular

```python
# Identificar valores en la columna genre

df['genre'].unique()
```

```python
# Calcular el promedio de popularidad para cada género y ordenarlo de forma descendente

promedios_por_genero = df.groupby('genre')['popularity'].mean().sort_values(ascending = False)
print(promedios_por_genero)
```

```python
# Gráfica

plt.figure(figsize=(15,10))
plt.barh(promedios_por_genero.index, promedios_por_genero, color=plt.cm.inferno_r(promedios_por_genero/100))
plt.xlabel('Popularidad promedio')
plt.ylabel('Género musical')
plt.title('Popularidad según género musical')
```

### Mapa de calor

Cómo se relacionan las características de las canciones entre sí.

```python
plt.figure(figsize=(15,10))
plt.title("Correlación entre características de las canciones de Spotify")
sns.heatmap(df.corr(), annot=True, cmap='inferno')
plt.show()
```

* Alta correlación entre energía y sonoridad
* Considerable correlación entre bailabilidad y valencia
* Considerable correlación entre popularidad, sonoridad, energía y bailabilidad

# Duración promedio de las canciones por género

## Medidas descriptivas según duración de tiempo de los géneros más populares

#### Pop

```python
pop_descripcion = df[df['genre'] == 'Pop']['duration_ms'].describe() # valores en milisegundos para el género Pop
print(pop_descripcion/1000) # valores en segundos
```

#### Rap

```python
rap_descripcion = df[df['genre'] == 'Rap']['duration_ms'].describe() # valores en milisegundos para el género Rap
print(rap_descripcion/1000)
```

#### Rock

```python
rock_descripcion = df[df['genre'] == 'Rock']['duration_ms'].describe() # valores en milisegundos para el género Rock
print(rock_descripcion/1000)
```

# Características de las canciones de género Pop, Rock y Rap más populares (popularidad mayor a 80%)

## Pop

```python
pop_sobre_ochenta = df.loc[(df['popularity'] > 80) & (df['genre'] == 'Pop')]
pop_sobre_ochenta['popularity'].sort_values(ascending=False)
print(pop_sobre_ochenta)
```

```python
pop_sobre_ochenta.describe()
```

```python
plt.figure(figsize=(15,10))
plt.title("Correlación entre características de las canciones de género Pop con más de 80% de popularidad")
sns.heatmap(pop_sobre_ochenta.corr(), annot=True, cmap='inferno')
plt.show()
```
